```python
import pandas as pd
```

```python
cosmo_df = pd.read_csv('tests/data/SolProp_v1.2_Data/CombiSolvH-QM.txt')
cosmo_df = cosmo_df[['smiles solvent','smiles solute','dHsolv (298K) [kcal/mol]']]
cosmo_df
```

```python
expt_df = pd.read_excel('tests/data/SolProp_v1.2_Data/CombiSolvH-Exp.xlsx')
expt_df = expt_df[['smiles_solvent','smiles_solute','dHsolv_avg [kcal/mol]']]
expt_df
```

```python
cosmo_df['smiles solvent'].value_counts()
```

```python
dict(expt_df['smiles_solvent'].value_counts())
```

```python
# select rows from cosmo_df where the smiles solvent is the most common solvent from the expt_df
reduced_cosmo_df = cosmo_df.loc[cosmo_df['smiles solvent'] == max(dict(expt_df['smiles_solvent'].value_counts()), key=dict(expt_df['smiles_solvent'].value_counts()).get)]
reduced_cosmo_df
```

```python
reduced_cosmo_df = reduced_cosmo_df[['smiles solute','dHsolv (298K) [kcal/mol]']]
reduced_cosmo_df = reduced_cosmo_df.rename(columns={'smiles solute':'smiles_solute','dHsolv (298K) [kcal/mol]':'dHsolv_cosmo'})
reduced_cosmo_df
```

```python
reduced_expt_df = expt_df.loc[expt_df['smiles_solvent'] == max(dict(expt_df['smiles_solvent'].value_counts()), key=dict(expt_df['smiles_solvent'].value_counts()).get)]
reduced_expt_df
```

```python
reduced_expt_df = reduced_expt_df[['smiles_solute','dHsolv_avg [kcal/mol]']]
reduced_expt_df = reduced_expt_df.rename(columns={'dHsolv_avg [kcal/mol]':'dHsolv_expt'})
reduced_expt_df
```

```python
# merge the two dataframes
merged_df = pd.merge(reduced_cosmo_df, reduced_expt_df, on='smiles_solute', how='outer')
merged_df.rename(columns={'smiles_solute':'smiles'}, inplace=True)
merged_df
```

```python
merged_df.to_csv('tests/data/SolProp_v1.2_Data/dHsolv_cosmo_expt.csv', index=False)
```
